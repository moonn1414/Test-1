```python
import pandas as pd
import matplotlib as mp
import seaborn as sb
import plotly.express as pl
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
# Survived, Pclass, Age, Sex. Parch
sb.countplot(data=data, x='Survived')
mp.pyplot.title('Dead_or_Alive')
```

```python
sb.countplot(data=data, x='Pclass')
mp.pyplot.title('Passplus')
```

```python
sb.histplot(data=data, x='Age')
mp.pyplot.title('AvgAge')
```

```python
sb.countplot(data=data, x='Sex')
mp.pyplot.title('HaveSex?')
```

```python
sb.countplot(data=data, x='Parch')
mp.pyplot.title('How_many_child')
```

```python
sb.boxplot(data=data, x='Age')
mp.pyplot.title('CvartilGraph')
#        28 
#       20  38 
#        64 
#        1   0  20, 2  20  38, 3  38  64
```

```python
sur=data['Survived'].value_counts()

mp.pyplot.pie(sur,labels=sur.index,autopct='%1.1f%%')
mp.pyplot.title('Dead_or_Alive')
```

```python
p=data['Pclass'].value_counts()
mp.pyplot.pie(p,labels=p.index,autopct='%1.1f%%')
mp.pyplot.title('Dead_or_Alive_Percent')
```

```python
cd=data[['Survived','Pclass','Age','SibSp','Parch','Fare']]
sb.pairplot(cd)
mp.pyplot.title('Kucha_info')
```

```python
sb=data[['Sex','Pclass']]
sbg=sb.groupby(['Sex','Pclass']).size().reset_index(name='size')
graph=pl.sunburst(sbg,path=['Pclass' ,'Sex'],values='size',title='Circleinfo')
graph.show()
```
